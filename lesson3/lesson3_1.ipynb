```python
# 字串和變數的整合
# 字串str 和 變數int 因為型別(type)不同無法相加運算，要用str將變數進行轉換
apple = 5
orange = 9

print("蘋果=" + str(apple) + "個,橘子=" + str(orange) + "個")
```

```python
# str template
# 字串插補變數
# 字串和變數整合,傳出1個全新的字串
print(f"蘋果={apple}個,橘子={orange}個")
```

```python
# 字串和變數的整合
apple = int(input("請輸入蘋果的數量:"))
orange = int(input("請輸入橘子的數量:"))

print("蘋果=" + str(apple) + "個,橘子=" + str(orange) + "個")
```

```python
#程式流程控制
#判斷
#關係(比較)運算子： <, <=, >, >+, ==(兩邊是否相等), !=(兩邊是否不相等)
age = 19

if age >= 18:
    print("滿18了")
    print("可以考駕照")
else:
    print("未滿18")
    print("不可以考駕照")
```

```python
#boolean布林值：只有true和false
# if判斷式，傳出布林值
# if+ 比較運算式 ->單項 傳出值為true才執行
# if+ 比較運算式 +else ->雙向 傳出值若為true,執行if部分；傳出值若為false,執行else部分。
```

```python
#巢狀判斷 
age = 39
if age < 18:
    print("未滿18")
    print("青少年")
else:
    if age < 40:
        print("青年")
    else:
        print("壯年")
```

```python
#多項選擇
#elif 將 if和else合併
age = 50
if age < 18:
    print("未滿18")
    print("青少年")
elif age < 40:
        print("青年")
else:
        print("壯年")
```

```python
#練習用多項選進行級距劃分
score=82
if score>=90:
    print("優")
elif score>=80:
    print("甲")
elif score>=70:
    print("乙")
elif score>=60:
    print("丙")
else:
    print("丁")
```

```python
#邏輯運算子 and 和 or
#and:嚴格得到true → and 兩側的運算元都是true, 該行才會傳出true
#or:容易得到true → or 兩側的運算元有一個是true, 該行就會得到true

#將and放到程式中增加可讀性(但沒有必要，不寫程式也能正常運算)

#python可以用數學方式敘述：a>b>c

score = 85
if score >= 90:
    print("優")
elif 90 > score >= 80:
    print("甲")
elif score < 80 and score >= 70:
    print("乙")
elif score < 70 and score >= 60:
    print("丙")
else:
    print("丁")
```

```python

#工具箱random
#randint(a,b)回傳隨機值介於ab之間 a<=隨機值<=b
#while迴圈，在不確定執行次數時使用
#while+布林值運算式 ：若傳出值為真，執行指令程式碼，並再次執行布林值運算式，直到傳出值為False，跳出迴圈。

import random


while random.randint(1,10) != 5:
    print("這不是5")

print("這是5")
```

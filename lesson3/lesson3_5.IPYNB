```python
nums = [1, 2, 3, 4, 5, 6]

#nums的索引位置編由左起依序為0~6 ；由右起依序為-1~-6
#nums[a:b}，a包還在要取出的數值中，b則不包含

#透過索引取值
print(nums[0])
print(nums[5])

#透過slice取值
print(nums[0:3])
print(nums[:3])
print(nums[-3:-1])
print(nums[-3:])

for num in nums:
    print(num)
```

```python
print(type(nums))
```

```python
#append 加入最後一個數字
nums.append(7)
```

```python
#insert(x,a) 在x位置插入數字a
nums.insert(0,-1)
```

```python
for num in nums:
    print(num)
```

```python
#clear清除
nums.clear()
nums
```

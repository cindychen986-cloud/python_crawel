```python
import time 

def download_file(file_name):
    print(f"開始下載: {file_name}")
    time.sleep(2)  # 模擬下載時間
    print(f"{file_name} 下載完成") 

def main_sync():
    start_time=time.time()
    download_file("檔案1.pdf")  
    download_file("檔案2.pdf")
    download_file("檔案3.pdf")
    end_time=time.time()
    print(f"執行時間: {(end_time - start_time):.2f} 秒") #.2f表示顯示小數點後兩位

if __name__ == "__main__":
    main_sync()

#download_file()中：檔案1.pdf為引數 ；file_name為參數，會呼叫main_sync裡面的download_file的引數
```

```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

async def download_file(file_name):
    print(f"開始下載: {file_name}")
    time.sleep(2)  # 模擬下載時間
    print(f"{file_name} 下載完成")
    
async def main_async():
    start_time=time.time()
    #await一次只丟一個任務，所以仍然是一步一步執行
    await download_file("檔案1.pdf")  
    await download_file("檔案2.pdf")
    await download_file("檔案3.pdf")
    end_time=time.time()
    print(f"執行時間: {(end_time - start_time):.2f} 秒") #.2f表示顯示小數點後兩位

if __name__ == "__main__":
    asyncio.run(main_async())
# async def 代表非同步函式
# await後面要接 async的function
# await 會將它後面的function丟去even loop
```

```python
#重點觀念 - await 的運作機制
# await + 非同步函式：await 用於呼叫非同步函式（async def 定義的函式），讓程式在等待該函式完成時，能夠讓出控制權給事件迴圈處理其他任務。
# await + 協程物件：協程物件（Coroutine）會被加入事件迴圈（Event Loop）中排隊執行。await 會暫停當前函式，直到該協程完成。
# await 的等待行為：await 會等待事件迴圈內的協程工作完成後，才繼續執行下一行程式碼。協程可以傳回值（例如 return "data"），也可以不傳回值（僅執行任務）。



#使用 asyncio.gather 同時啟動多個任務。
import asyncio
import nest_asyncio
nest_asyncio.apply()

async def download_file(file_name):
    print(f"開始下載: {file_name}")
    await asyncio.sleep(2)  # 模擬下載文件所需的時間
    print(f"{file_name} 下載完成")
    
async def main_async():
    start_time=time.time()
    # 同時啟動 3 個下載任務
    await asyncio.gather(
        download_file("檔案A.pdf"),
        download_file("檔案B.pdf"),
        download_file("檔案C.pdf")
    )
    end_time=time.time()
    print(f"執行時間: {(end_time - start_time):.2f} 秒") #.2f表示顯示小數點後兩位

if __name__ == "__main__":
    asyncio.run(main_async())
```
